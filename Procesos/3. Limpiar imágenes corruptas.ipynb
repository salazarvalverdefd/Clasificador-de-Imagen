```python
#eliminar corruptos
from glob import glob
import os
def ls(expr = '*.*'):
    return glob(expr)

#carpeta = '2. Dataset_Imágenes_No Vacías'
carpeta = '1. Dataset_Imágenes_Original'
archivos=ls(carpeta+'/*/*/*.jpg')
total=0
eliminados=0

for imagen in archivos:
    #print(imagen)
    sizefile = os.path.getsize(imagen)
    total=total+1
    
    if sizefile < 50:
        os.remove(imagen)
        #print(sizefile)
        eliminados=eliminados+1
print(total)
print(eliminados)
    
```

```python
##### OTROS #####
#juntar imágenes
#30 etiquetas
from glob import glob
import os
import shutil

def ls(expr = '*.*'):
    return glob(expr)

carpeta = '2. Dataset_Imágenes_No Vacías - copia (3)'
subcarpeta='testSet'
archivos=ls(carpeta+'/'+subcarpeta+'/*')
total=0
remover=0
for carpeta_imagen in archivos:
    total=total+1
    archivos_img=ls(carpeta_imagen+'/*')
    #print(carpeta_imagen)
    c=carpeta_imagen.replace(carpeta+'/'+subcarpeta+'\\','')
    #print(c)
    a,b=c.split('_',1)
    directorio=carpeta+'/'+subcarpeta+'\\'+a
    name=a
    if carpeta_imagen.find('Otros_')>=0:
        directorio=carpeta+'/'+subcarpeta+'\\'+b
        name=b
    print(directorio)
    
    for img in archivos_img:
        remover=remover+1
        #print(img)
        nombre=name+'-'+str(remover)
        try:
            os.stat(directorio)
        except:
            os.makedirs(directorio)
        #print(nombre)
        shutil.copy(img, directorio+'\\'+nombre+'jpg')
        
print(total)
print(remover)
    
```
